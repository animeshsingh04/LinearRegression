```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
def readfile(filename):
    data = pd.read_csv(filename)
    X = data.iloc[:,-1].values.reshape(-1,1)
    Y = data.iloc[:, 1].values.reshape(-1,1)
    return X, Y
```

```python
# Building the model with least square root method
def buildModel(X,Y):
    X_mean = np.mean(X)
    Y_mean = np.mean(Y)
    num = 0
    den = 0
    for i in range(len(X)):
        num += (X[i] - X_mean)*(Y[i] - Y_mean)
        den += (X[i] - X_mean)**2
    m = num / den
    c = Y_mean - m*X_mean
    return m,c
```

```python
# Making predictions
def bestFitLine(X,m,c):
    Y_pred = m*X + c
    return Y_pred
```

```python
#Plot graph
def plotGraph(X,Y,Y_pred):
    plt.scatter(X, Y) # actual
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
    plt.show()
    return
```

```python
def predict(x, m,c):
    y_pred = c + m*x
    return y_pred
```

```python
# load the dataset
X, Y = readfile('salary_data.csv')

# build the model
m,c = buildModel(X,Y)

Y_pred=bestFitLine(X,m,c)
    
df_result = pd.DataFrame({'Actual': Y.flatten(), 'Predicted': Y_pred.flatten()})
df_result
    
    
```
