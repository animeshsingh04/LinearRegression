```python
# data manipulation

import numpy as np
import pandas as pd

from sklearn import svm
from random import randint
# plotting

import matplotlib.pyplot as plt
import matplotlib.dates as mdate

# machine learning
from sklearn.datasets.samples_generator import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from datetime import datetime as dt
from sklearn.cluster import KMeans

# data fetching

from pandas_datareader import data as pdr

```

```python
df = pd.read_csv("data_k.csv")  # data reading from csv folder
df1 = []
df2=[]
#print(df)
for row in df['date']:  # separating day from time form
    df1.append(row[0:2])
x = df1
y1 = df['open']
y2 = df['high']
y3 = df['close']

#print(y1)

df2=df['open'].values #storing open price into array
tp = np.array(df2)
fp = np.array(x)
#print(df2)
print (fp)


```

```python
#kmeans on open price
plt.title('Dataset for open price')
plt.scatter(x, y1, alpha=1)
#plt.show()
X1 = []
X1=np.column_stack((x,y1))
plt.scatter(x,y1)
kmeans1 = KMeans(n_clusters=3, random_state=0).fit(X1)
#print(kmeans1)
y1_kmeans = kmeans1.predict(X1)
plt.title('cluster for open price')
plt.scatter(x, y1, c=y1_kmeans, cmap='viridis')
centers1 = kmeans1.cluster_centers_
#print(centers1)
plt.scatter(centers1[:, 0], centers1[:, 1], c='black' ,alpha=0.5);
plt.show()
```

```python
#kemeans for high price
plt.title('Dataset for high price')
plt.scatter(x, y2, alpha=1)
#plt.show()
X2 = []
X2=np.column_stack((x,y2))
plt.scatter(x,y2)
kmeans2 = KMeans(n_clusters=3, random_state=0).fit(X2)
#print(kmeans2)
y2_kmeans = kmeans2.predict(X2)
plt.title('Cluster for high price')
plt.scatter(x, y2, c=y2_kmeans, cmap='viridis')
centers2 = kmeans2.cluster_centers_
#print(centers2)
plt.scatter(centers2[:, 0], centers2[:, 1], c='black' ,alpha=0.5);
plt.show()
```

```python
#kmeans close price
plt.title('Dataset for close price ')
plt.scatter(x, y3, alpha=1)
#plt.show()
X3 = []
X3=np.column_stack((x,y3))
plt.scatter(x,y3)
kmeans3 = KMeans(n_clusters=3, random_state=0).fit(X3)
#print(kmeans3)
y3_kmeans = kmeans3.predict(X3)
plt.title('cluster for close price ')
plt.scatter(x, y3, c=y3_kmeans, cmap='viridis')
centers3 = kmeans3.cluster_centers_
#print(y3_kmeans)
plt.scatter(centers3[:, 0], centers3[:, 1], c='black' ,alpha=0.5);
plt.show()
```
